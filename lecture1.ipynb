## Practice: Matrix multiplication

```python
def matrix_multiplication(a, b):
  row_a = len(a)
  col_a = len(a[0])
  row_b = len(b)
  col_b = len(b[0])

  if (col_a != row_b):
    return "Error: Number of columns of the multiplicand should be the same as the number of rows of the multiplier"

  result = [[0] * col_b for i in range(row_a)]

  for i in range(row_a):
    for j in range(col_b):
      sum = 0
      for k in range(col_a):
        sum += a[i][k] * b[k][j]
      result[i][j] = sum
  return result

a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
b = [
    [10, 11, 12],
    [13, 14, 15],
    [16, 17, 18]
]

matrix_multiplication(a, b)
```

## Autograd

```python
x = torch.tensor(4.0, requires_grad=True)
y = x**2 + 5
y.backward()
print(x.grad)
```

```python
x = torch.tensor(4.0, requires_grad=True)
y = torch.abs(x - 4)
y.backward()
print(x.grad)
```

```python
m = torch.tensor(2., requires_grad=True)
x = torch.tensor(4., requires_grad=True)
b = torch.tensor(3., requires_grad=True)
```

```python
x.requires_grad_(False)
```

```python
y = m * x + b
y.backward()
```

```python
print(m.grad)
print(x.grad)
print(b.grad)
```

# Ipywidget

```python
from ipywidgets import interact

@interact(a=1.5, b=1.5, c=1.5)
def plot(a, b, c):
    x = torch.linspace(-10, 10, 100)
    y = a * x**2 + b * x + c
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.plot(x, y)
    plt.show()
```

```python
@interact(start=5, end=10, step=1)
def print_range(start, end):
    print(torch.arange(start, end))
```
