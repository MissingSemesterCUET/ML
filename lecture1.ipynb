```python
def matrix_multiplication(a, b):
  row_a = len(a)
  col_a = len(a[0])
  row_b = len(b)
  col_b = len(b[0])

  if (col_a != row_b):
    return "Error: Number of columns of the multiplicand should be the same as the number of rows of the multiplier"

  result = [[0] * col_b for i in range(row_a)]

  for i in range(row_a):
    for j in range(col_b):
      sum = 0
      for k in range(col_a):
        sum += a[i][k] * b[k][j]
      result[i][j] = sum
  return result

a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
b = [
    [10, 11, 12],
    [13, 14, 15],
    [16, 17, 18]
]

matrix_multiplication(a, b)
```
