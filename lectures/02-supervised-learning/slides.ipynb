## What is Supervised Learning?

- Training with **labeled data**.
- The goal is to learn a mapping $f(x) = y$.
- $x$ = features, $y$ = target/label.

---

## Linear Regression

- Predicting continuous values.
- Formula: $y = mx + b$

## Logistic Regression

- Binary classification (0 or 1).
- Uses the Sigmoid function to map predictions to probabilities.

## Decision Trees

- Flowchart-like structure.
- Easy to interpret and visualize.
- Great for handling non-linear data.

---

## Evaluation Metrics

- **Accuracy**: Correct / Total
- **Precision**: TP / (TP + FP)
- **Recall**: TP / (TP + FN)
