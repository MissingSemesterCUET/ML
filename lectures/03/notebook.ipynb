### Two Main Reasons We Visualize Data

#### 1. Exploratory Visualization (Finding Insights)

Used when **you don't yet know what the data is telling you**.

**Example:**  
A teacher collects exam scores from 500 students and creates several graphs to see:

- Which subjects students struggle with
- Whether scores improved over time
- Whether there are unusual results

The goal is to **explore and learn from the data**.

> **Exploratory Visualization = "What can I discover from this data?"**

---

#### 2. Explanatory Visualization (Communicating Insights)

Used when **you already know the insight and want to show it to others**.

**Example:**  
After analyzing the scores, the teacher finds that students perform much better in Mathematics than in Physics. They create a clean chart and present it to the class.

The goal is to **explain a specific message clearly**.

> **Explanatory Visualization = "How can I help others understand this insight?"**

### Importing Matplotlib

```python
import matplotlib.pyplot as plt
# pyplot is a sub-module that provides a simple way to create graphs quickly; plt is the alias commonly used for this library
```

# Line plots

```python
x = [1,2,3]
y = [4,5,6]

plt.plot(x,y) # tells Matplotlib to connect these points with a line --> (1,4),(2,5),(3,6)
plt.show() # shows the graph on screen
```

```python
# To get a more clear idea, we can turn on grids as well
plt.plot(x,y)
plt.grid()
plt.show()
```

Does this graph actually tell us anything meaningful? Right now, it doesn’t. Let’s make it clear by adding labels and markers
> Markers are symbols that show the exact data points on a graph.

| Marker | Name        | Description              | Example Use |
|--------|------------|--------------------------|--------------|
| `o`    | Circle     | Most commonly used point | General data points |
| `s`    | Square     | Sharp, block-style point | Categorical data |
| `^`    | Triangle Up| Directional marker       | Trends or signals |
| `v`    | Triangle Down | Inverse direction marker | Decreasing trends |
| `x`    | Cross      | Highlighted point        | Errors or special points |
| `+`    | Plus       | Simple cross marker      | Comparisons |
| `*`    | Star       | Emphasis marker          | Important values |
| `D`    | Diamond    | Strong visual emphasis   | Key data points |

```python
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot")
plt.plot(x,y, marker = 'o',linestyle='--') # plt.plot(x, y, 'o--') --> shorthand!
plt.show()
```

Now that we’ve seen real data, let’s explore the different ways we can visualize it.

```python
# Dummy data of study hours and marks obtained by different students
students = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
gender = ["M","F","M","F","F","M","F","M","F","M"]
hours = [1, 2, 2, 3, 4, 5, 6, 6, 7, 8]
marks = [25, 35, 40, 55, 60, 62, 70, 75, 85, 95]
```

# Scatter plot - relationship

Think about questions to analyse particular trends. For e.g:
> Do study hours affect marks?

```python
plt.scatter(hours, marks,color="red",s=50, alpha=0.5) # s controls the size of the points, alpha controls the transparency
plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

# Bar charts and Histogram - comparison
> A bar chart is used when we want to compare values across categories

```python
plt.bar(students, marks)
plt.title("Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

> A histogram does NOT show individual students. Instead, it shows how the data is distributed.

| Bin Range | Values Inside | Count |
| --------- | ------------- | ----- |
| 25 – 39   | 25, 35        | 2     |
| 39 – 53   | 40            | 1     |
| 53 – 67   | 55, 60, 62    | 3     |
| 67 – 81   | 70, 75        | 2     |
| 81 – 95   | 85, 95        | 2     |



```python
plt.hist(marks, bins=5) # marks = [25, 35, 40, 55, 60, 62, 70, 75, 85, 95]

# bins = number of equal intervals we divide the data into
# range = (max - min) = (95 - 25) = 70
# class interval (approx bin width) = 70 / 5 = 14 ; this is the class interval

plt.title("Distribution of Marks")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()
```

# Boxplot

# Colormaps
Our brain understands patterns and colors much faster than raw values.

That’s where colormaps come in.

A colormap is a rule that maps numerical values to colors. Instead of seeing just numbers, we see a visual gradient that helps us quickly understand how values change across a dataset.

For example, low values might appear as dark colors, and high values as bright colors. This makes it easier to instantly spot patterns like clusters, trends, or anomalies.

```python
import numpy as np

marks = np.array([25, 35, 40, 55, 60, 62, 70, 75, 85, 95])
data = marks.reshape(2, 5)
plt.imshow(data, cmap='viridis') # turning numbers into a picture.
plt.colorbar()
plt.title("Colormap Example")
plt.show()

# Each box is one student.
# The number is their mark.
# The color-darker means low marks, brighter means high marks.

# [[25, 35, 40, 55, 60],
#  [62, 70, 75, 85, 95]]
```

# Subplots
Subplots let us place multiple plots in one figure.

```python
x = [1, 2, 3, 4]

# y1 = first trend (e.g., Student A marks over time)
y1 = [10, 20, 15, 25]

# y2 = second trend (e.g., Student B marks over time)
y2 = [20, 15, 30, 14]

# a figure with 1 row and 2 columns
# fig = whole canvas
# ax = individual plots inside the canvas
fig, ax = plt.subplots(1, 2,figsize=(5,2)) # width,height

# ax[0] means first plot (left side)
ax[0].plot(x, y1)              # draw line graph for y1
ax[0].set_title("Trend A")     # title for first graph

ax[1].plot(x, y2)              # draw line graph for y2
ax[1].set_title("Trend B")     # title for second graph

# adjust spacing so plots don't overlap
plt.tight_layout()

# save first before showing as python may clear memory after displaying;
# so calling save function after display may cause an empty figure to be saved
plt.savefig("my_plot.png")
plt.show()
```

# Introduction to Seaborn — why use it over raw Matplotlib?

Seaborn is a Python visualization library built on top of Matplotlib. It is used for statistical data visualization

# Why Seaborn?
- Less code compared to Matplotlib
- Better default styles (clean plots without manual styling)
- Works directly with Pandas DataFrames
- Easier statistical visualization (distribution, relationships, categories)

| Feature | Matplotlib | Seaborn |
|----------|------------|---------|
| Level | Low-level plotting library | High-level statistical visualization library |
| Input style | Works with lists/arrays | Works directly with Pandas DataFrames |
| Code length | More verbose (more steps) | Shorter and simpler |
| Default appearance | Basic, needs manual styling | Pre-styled with better defaults |
| Data grouping | Manual | Automatic using column names |

# Histogram

```python
import seaborn as sns
```

```python
data = [10, 20, 20, 30, 30, 30, 40, 50]

sns.histplot(data, bins=5)
# More bins = more detailed view, fewer bins = simpler view

plt.title("Histogram - Distribution of Data")
plt.show()
```

What is wrong with histogram?
- Varying the number of bins will result in a different histogram
- This results in different figures of the same data

```python
sns.histplot(data, bins=10)
plt.title("Distribution of data")
plt.show()
```

**so, histogram has inconsistent presentation**

# KDE Plot - Kernel Density Estimate

Imagine drawing a smooth hill over the histogram bars — that hill is the KDE.
- Higher curve = data is more concentrated around that region.

<img src ="https://i.imgur.com/Yrsk0HU.png" width="300" height="200">



- Bandwidth : size of kernel so if bandwidth is low , then the kernel will be narrower and taller and when is it high then the kernel will be wider and shorter

- This determines how far the influence of the points reaches

- Kernel function determines the shape of the kernel, we mostly use Gaussian kernel which has a bell shaped curve


```python
kde_data = [1, 3, 3.5, 4, 6, 8]
sns.kdeplot(kde_data)
```

```python
sns.kdeplot(kde_data, fill = True)
```

```python
sns.kdeplot(kde_data,bw_adjust=0.5) # decreasing the bw will cause more jagged hills, while increasing it will result in a uniform curve, with less variance
```

# Categorical Plots

```python
marks = [25, 35, 40, 55, 60, 62, 70, 75, 85, 180]

sns.boxplot(y=marks)

# Box represents middle 50% of data (Q1 to Q3)
# Line inside box = median
# Whiskers = min and max range (excluding outliers)
```

### Violin Plot : Is data single-peaked or multi-peaked?
A violin plot is a boxplot with a KDE curve attached on both sides. The width at any point tells you how many values are concentrated there
- wider = more data at that value.

```python
# sns.violinplot()
# Combines boxplot + distribution shape
# Shows density of data (wider = more values in that range)

sns.violinplot(y=marks)

plt.title("Violin Plot - Distribution Shape + Summary")
plt.show()
```

### Swarmplot
- What exactly are the raw values?

```python
# Shows each individual data point clearly
# Useful when dataset is small

sns.swarmplot(y=marks)

plt.title("Swarm Plot - Individual Data Points")
plt.show()
```

# Relational Plots

```python
# sns.scatterplot()
# Used to show relationship between two variables

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

sns.scatterplot(x=x, y=y)

plt.title("Scatter Plot - Relationship Between Variables")
plt.show()

# Each dot = one data point
# Used to see correlation or pattern
```

```python
# sns.lineplot()
# Used to show trend over time or ordered data

sns.lineplot(x=x, y=y)

plt.title("Line Plot - Trend Over Time")
plt.show()

```

### Heatmap

```python
import numpy as np

data = np.array([
    [1, 0.8, 0.2],
    [0.8, 1, 0.5],
    [0.2, 0.5, 1]
])

sns.heatmap(data, annot=True)

plt.title("Heatmap - Relationship Strength")
plt.show()

# annot=True shows numbers inside cells
# Darker color = stronger relationship
```

### Pairplot : How does every feature relate to every other feature?
Takes a DataFrame and automatically makes scatter plots for every pair of columns. The diagonal shows a histogram (or KDE) of each single variable. It gives a bird's-eye view of your whole dataset at once.

```python
# sns.pairplot()
# Used to show relationships between multiple variables

import pandas as pd

df = pd.DataFrame({
    "math": [60, 70, 80, 90],
    "science": [65, 75, 85, 95],
    "english": [55, 60, 70, 80]
})

sns.pairplot(df)

plt.show()

# Creates scatter plots for all variable pairs
# Helps understand dataset quickly
```

```python
df = pd.DataFrame({
    "math": [60, 70, 80, 90, 40, 45, 50, 55],
    "science": [40, 75, 85, 95, 42, 48, 52, 88],
    "english": [55, 60, 70, 80, 35, 40, 45, 50],
    "type": ["good", "good", "good", "good",
             "weak", "weak", "weak", "weak"]
})

sns.pairplot(
    df,
    hue="type", # Color points based on a category column
    diag_kind="kde",
    plot_kws={"alpha": 0.7, "s": 60}
)
```

- strange clusters in scatter = incorrect data grouping
