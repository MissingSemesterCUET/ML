## 1. Creating data
Pandas gives us **two things**:

- **DataFrame** - a full table, with rows and columns. Like an Excel sheet.
- **Series** - just **one column** of that table.

**Simple way to think:** a DataFrame is a table. A Series is one column of it.
Glue some Series together side by side -> you get a DataFrame.

### Creating a DataFrame

### From a dictionary

- **Dictionary -> DataFrame**
  - **Keys** become **column names**
  - **Values** become **column data** (lists, arrays, or Series)

````python
import pandas as pd

# a datafram from a dictionary: keys become column names, values become the column data
flood = pd.DataFrame({
    'district': ['Sylhet', 'Feni', 'Kurigram', 'Noakhali'],
    'rain':     [320, 350, 280, 330],
    'families': [12000, 15200, 9800, 13100]
})
flood
````

See the `0, 1, 2, 3` on the left? Pandas numbers every row by itself.
That left side is called the **index**. We can put our own labels there instead:

````python
pd.DataFrame(
    {'shelter': [45, 52],
     'volunteers': [340, 300]},
    index = ['Sylhet', 'Feni']
)
````

### From a list of lists

A DataFrame can also be made from a list of lists. Each inner list is **one row**

````python
pd.DataFrame(
    [
        ['Sylhet', 320],
        ['Feni', 350],
        ['Kurigram', 280]
    ],
    columns=['district', 'rain'],   # column names
)
````

And one single column alone = a **Series**.
### from list


````python
shelters = pd.Series([45, 52, 38], index=['Sylhet', 'Feni', 'Kurigram'])
shelters
````

- **Dictionary -> Series**
  - **Keys** become the **index**
  - **Values** become the **data**

````python
rainfall = {
    'Sylhet': 320,
    'Feni': 350,
    'Kurigram': 280,
    'Noakhali': 330
}

rain = pd.Series(rainfall)

rain
````

more ways to make a Series from a **numpy array**:

````python
import numpy as np

rain_arr = np.array([320, 410, 280])
pd.Series(rain_arr, index=['Sylhet', 'Sunamganj', 'Kurigram'])
````

**masking a Series** : write a True/False test inside `[ ]` and only the passing values stay:

````python
shelters[shelters > 40]
````

## 2. Reading and saving files

nobody types data by hand in real work. data comes as a **file**-usually a **CSV** (comma seperated values)file.

for example, district office sends us this year's flood report. We'll make that file
once, then read it back — because reading files is how every real project starts.

````python
report = pd.DataFrame({
    'district': ['Sylhet', 'Sunamganj', 'Kurigram', 'Feni',
                 'Netrokona', 'Lalmonirhat', 'Habiganj', 'Noakhali'],
    'division': ['Sylhet', 'Sylhet', 'Rangpur', 'Chattogram',
                 'Mymensingh', 'Rangpur', 'Sylhet', 'Chattogram'],
    'rain':     [320, 410, 280, 350, 290, 260, 300, 330],
    'water':    [850, 920, 780, 810, 760, None, 800, 790],  # water level; Lalmonirhat's meter broke!
    'families': [12000, 18500, 9800, 15200, 7600, 8900, 10400, 13100],
    'shelters': [45, 60, 38, 52, 30, 33, 41, 48]
})
report.to_csv('flood_report.csv', index=False) # save to a file
flood = pd.read_csv('flood_report.csv')         # read from a file
flood
````

## 3. picking data from table
 the skills you will use every single dat. three ways:



 1. `take one column -> df['col']`


````python
flood['district']
````

2. pick by position(0, 1, 2...) -> `df.iloc[...]`

iloc integer location

````python
flood.iloc[0]
````

3. pick by name -> `df.loc[...]`

loc mean location by label

`loc[rows, columns]`

````python
flood.loc[:, ['district', 'families']]   # all rows, only these two columns
````

````python
# Selecting a single row
print(flood.loc[0])

# Selecting a single cell
display(flood.loc[0, 'rain'])
# Example 3: Selecting a range of rows (from index 1 to 3, inclusive) and all columns
display(flood.loc[1:3, :])
print(flood.loc[1:3, 'division':'families'])

# Selecting specific rows and specific columns
display(flood.loc[[0, 2, 5], ['district', 'families', 'shelters']])
````

**A trap everyone falls in once:** `iloc[0:3]` gives rows 0, 1, 2 - it stops before 3,
just like Python lists. But `loc[0:3]` gives rows 0, 1, 2 **and 3**. `loc` includes both ends.

### map -> chnage every value in a column at once

Take a value ->
Apply the function ->
Return the result

````python
# how much above or below avg rain did each district get???
avg = flood['rain'].mean()
flood['rain'].map(lambda r: r - avg)
````

lambda:
```python
def subtract_average(r):
    return r - avg

````python
# apply() is the same idea but it can look at the whole row at once
def danger(row):
    return row['rain'] / 100 + row['families'] / 1000

flood.apply(danger, axis=1)   # axis=1 means go row by row
````

## 7. joining tables

### Two ways to join

- **`pd.concat()`** - put tables **on top of each other** (same columns, more rows).
- **`pd.merge()`** - join two tables **side by side** using a common column(same entities, more columns).

````python
# concat: two districts sent their report late, add them below
late = pd.DataFrame({
    'district': ['Bandarban', 'Chandpur'],
    'division': ['Chattogram', 'Chattogram'],
    'rain': [370, 340],
    'water': [700, 720],
    'families': [5400, 8800],
    'shelters': [21, 35],
    'help_sent': [False, False],
    'per_shelter': [257.1, 251.4]
})
all_districts = pd.concat([flood, late])#  ignore_index=True
all_districts.tail(4)
````

````python
# merge: a different table has the relief stock of each district — join them
stock = pd.DataFrame({
    'district':  ['Sylhet', 'Sunamganj', 'Kurigram', 'Feni'],
    'rice_tons': [120, 95, 80, 110],
    'volunteers': [340, 280, 210, 300]
})
pd.merge(flood, stock, on='district')
````
