# Welcome to ML — What is it and why should you care?

Machine Learning (ML) is a field of computer science where computers learn patterns from data and improve their performance without being explicitly programmed for every rule.

Imagine you want a computer to recognize cats in pictures.
* **Task (T):** What the system should do
  → Example: Identify whether an image contains a cat or not

* **Experience (E):** The data/examples it learns from
  → Example: Thousands of labeled cat and non-cat images

* **Performance (P):** How well it is doing
  → Example: How accurately it identifies cats in new images

**Speech Recognition Example**

Person says:

*One*

*Two*[ high-pitch sound (“T”)]

hardcode an algorithm that measures high-pitch sound intensity
and use that to distinguish ones and twos; but obviously this technique will
not scale to thousands of words spoken by millions of very different people

# Real-World Applications of Machine Learning

### Vision • NLP • Recommendation Systems • Generative AI

* **Computer Vision**
  Enables computers to understand and analyze images or videos.
  *Examples:* cat classification which I have discussed earlier, detecting tumors in brain scans using CNN and transformer

* **Natural Language Processing (NLP)**
  Helps computers understand and generate human language.
  *Examples:* Automatically flagging offensive comments on discussion forums.

* **Recommendation Systems**
  Suggests content or products based on user behavior and preferences ( artificial neural network)
  *Examples:* Netflix movie recommendations, YouTube videos, Amazon products.

* **Generative AI**
  Creates new content such as text, images, music, or code.
  *Examples:* ChatGPT, AI image generators, coding assistants.

### Supervised Learning

### Unsupervised Learning

In unsupervised learning, the data has no labels.  
The model tries to discover hidden patterns or structures on its own.

> “The system learns without a teacher.”

---

***Common Tasks in Unsupervised Learning***

* **Clustering**
Groups similar data points together automatically.  

*Example:* Grouping website visitors based on interests and behavior.

clustering.png

---

* **Dimensionality Reduction**
Reduces complexity while keeping important information.  

*Example:* Compressing many related features into fewer meaningful ones.

---

* **Anomaly Detection**
Detects unusual or abnormal patterns in data.  

*Example:* Detecting fraudulent credit card transactions.

anomaly_detection.png

---

* **Association Rule Learning**
Finds relationships between items in large datasets.  

*Example:* Customers buying chips often also buy soft drinks.

### Reinforcement Learning

An agent learns by interacting with an environment.  
The agent takes actions, receives rewards or penalties, and gradually learns the best strategy.

# Control Flow + Loops

### `if / elif / else`

```python
food_packets = 3

if food_packets >= 5:
    print("Enough food for all cats.")

elif food_packets >= 2:
    print("Food is limited.")

else:
    print("Need more food urgently.")
```

### `for` loop

# Dictionaries

```python
cat = {
    "name": "Milo",
    "age": 2,
    "vaccinated": True
}

print(cat)
```

```python
# Accessing values
print(cat["name"])
print(cat["age"])
```

```python
# Updating values
cat["age"] = 3

print(cat)
```

```python
# Adding new key
cat["color"] = "Orange"

print(cat)
```

```python
# Looping through dictionary
for key, value in cat.items():
    print(key, value)
```

# Object Multiplication

```python
matrix = [[0] * 3] * 3

matrix[0][0] = 1

print(matrix)
```

> all rows changed!

```python
# correct way
matrix = [[0] * 3 for _ in range(3)]

matrix[0][0] = 1

print(matrix)
```

# List Comprehension

```python
food_needed = [2, 5, 1, 4]

urgent_cases = [food for food in food_needed if food < 2]

print(urgent_cases)
```

# Functions

```python
# function with argument
def feed(cat_name):
    print(f"Feeding {cat_name}")

feed("Milo")
```

```python
# Return value
def calculate_food(weight):
    return weight * 0.04

food = calculate_food(5)

print(food)
```

```python
# Multiple parameters
def rescue_summary(name, age):
    return f"{name} is {age} years old."

print(rescue_summary("Luna", 2))
```
