## What is Machine Learning?

- A subset of AI that allows systems to learn from data.
- Arthur Samuel (1959): "Field of study that gives computers the ability to learn without being explicitly programmed."
- Focus on statistical patterns and predictive modeling.

---

## Types of ML

1. **Supervised Learning**: Mapping inputs to outputs using labeled data.
2. **Unsupervised Learning**: Discovering structure in unlabeled data.
3. **Reinforcement Learning**: Learning via rewards and penalties.

---

## Applications in Real World

- **Image Recognition**: FaceID, medical imaging.
- **NLP**: Translation, sentiment analysis, chatbots.
- **Recommender Systems**: Netflix, YouTube, Spotify.
- **Autonomous Vehicles**: Self-driving cars.

---

## Why Python for ML?

Python has a rich ecosystem of libraries:
- **NumPy**: Numerical computing.
- **Pandas**: Data manipulation.
- **Scikit-Learn**: General ML tools.
- **PyTorch / TensorFlow**: Deep Learning.
