```python
class ndarray:
    def __init__(self, data):
        self.data = data
        self.shape = (len(data),)
    
    def __repr__(self):
        return f"array({self.data})"

    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        return self.data[index]

    def __setitem__(self, index, value):
        self.data[index] = value
    
    def __element_wise_operate(self, other, f):
        result = []
        if isinstance(other, ndarray):
            for i,j in zip(self.data, other.data):
                result.append(f(i,j))
        else:
            for item in self.data:
                result.append(f(item,other))
        return ndarray(result)
    
    def __add__(self, other):
        return self.__element_wise_operate(other, lambda x,y: x+y)
            
    def __mul__(self, other):
        return self.__element_wise_operate(other, lambda x,y: x*y)
    
    def __sub__(self, other):
        return self.__element_wise_operate(other, lambda x,y: x-y)
    
    def __truediv__(self, other):
        return self.__element_wise_operate(other, lambda x,y: x/y)

    def __floordiv__(self, other):
        return self.__element_wise_operate(other, lambda x,y: x//y)

    def sum(self):
        ans = 0
        for i in self.data:
            ans+=i
        return ans
    
    def prod(self):
        result = 1
        for item in self.data:
            result *= item
        return result
    
    def mean(self):
        return self.sum() / len(self.data)

    def min(self):
        result = self.data[0]
        for item in self.data:
            if item < result:
                result = item
        return result

    def max(self):
        result = self.data[0]
        for item in self.data:
            if item> result:
                result = item
        return result

def array(data):
    return ndarray(data)

a = array([1,2,3,4])
b = array([10,20,30,40])

a.prod()
```
