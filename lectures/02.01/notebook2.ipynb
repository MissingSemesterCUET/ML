# Row vs Column Vectors

```python
a = np.array([1,2,3,4])
a
```

```python
a.shape
```

This is not a row vector.

It is not a column vector.

It is just a 1D array.

If you want a column vector:

```python
a = a.reshape(-1, 1)
a
```

```python
a.shape
```

If you want a row vector

```python
a = a.reshape(1, -1)
a
```

```python
a.shape
```

This distinction becomes important for matrix multiplication.

# Vectors are 1D Matrices (our convention)

Mathematically, vectors are different objects.

But for deep learning, it's convenient to think of them as matrices with one row or one column.

Then the dot product is just matrix multiplication.

```python
a = np.array([1,2,3]).reshape(-1, 3)
a
```

```python
b = np.array([4,5,6]).reshape(3,-1)
b
```

```python
dot_product =a @ b
dot_product
```

```python
dot_product.item()
```

# Broadcasting

## Matrix + Scalar

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
A + 10
```

## Matrix + Row Vector

```python
v_row = np.array([[10, 20, 30]])
A + v_row
```

## Matrix + Column Vector

```python
v_col = np.array([
    [10],
    [20]
])
A + v_col
```

## Row Vector + Column Vector

```python
r = np.array([[1, 2, 3]])
c = np.array([
    [10],
    [20],
    [30]
])
r + c
```

## Broadcasting Rules

NumPy compares shapes element-wise starting with the rightmost dimensions and works its way left. Two dimensions are compatible when:
1. They are **equal**.
2. One of them is **1**.

If these conditions are not met, a `ValueError` is raised.

```python
# A + np.array([[10, 20], [30, 40]]) # Error: operands could not be broadcast together
```
