# Stage 0: Intro to OOP

I know many of you are not familiar with OOP. So, we will teach you the basics of OOP along the way.

The concept of OOP was created to make the lives of programmers easier. Theoretically, we can build build anything from scratch without ever knowing what OOP is. But that will be very very very difficult. With OOP, its much easier.

Altough it came to make our lives, easier, it's not taugh very well. So, students are often afraid of OOP.

Here is big idea of OOP:

When you start a new project, with a blank screen infront of you, with no code written yet, don't just start writing the code. Think about what are the **objects** you are going to work with.

Say you want to build a classroom management system for CUET.

The object for you are:
- students
- teachers
- courses
- classroom

After you have created the list of objects, think about what properties/attributes do the objects have.

For example:
- Attrs of student:
    - name
    - student id
    - student email
    - dept
    - batch
    - hall name
- Attrs of teacher:
    - name
    - teacher id
    - teacher email
    - dept
    - affiliation
- Attrs of course:
    - name
    - course id
    - list of course teachers
- Attrs of classroom
    - room number
    - building number
    - lab/class

Then think about how each of the objects **behave** (what they can do. we call them methods).

Example:
- Methods of student:
    - can enroll in course
    - can attend a class
- Methods of teacher:
    - can conduct a course
    - can take attendance
- Methods of course:
    - cant do anything
- Methods of classroom:
    - cant do anything

That's all OOP is
