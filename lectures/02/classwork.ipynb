```python
class ndarray:
    def __init__(self, data):
        self.data = data
        self.shape = (len(data),)
    
    def __repr__(self):
        return f"array({self.data})"

    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        return self.data[index]

    def __setitem__(self, index, value):
        self.data[index] = value
    
    def __add__(self, other):
        result = []
        for i,j in zip(self.data, other.data):
            result.append(i+j)
        return ndarray(result)
    
    

def array(data):
    return ndarray(data)

a = array([1,2,3,4])
b = array([10,20,30,40])

a*b
```

```python
class ndarray:
    def __init__(self, data):
        self.data = data
        self.shape = (len(data),)

    def __repr__(self):
        return f"array({self.data})"

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        return self.data[index]

    def __setitem__(self, index, value):
        self.data[index] = value

    def __add__(self, other):
        result = []
        for i, j in zip(self.data, other.data):
            result.append(i + j)
        return ndarray(result)

    def __mul__(self, other):
        result = []
        for i, j in zip(self.data, other.data):
            result.append(i * j)
        return ndarray(result)

def array(data):
    return ndarray(data)

a = array([1,2,3,4])
b = array([10,20,30,40])

print(a * b)
```
