# Maths for ML

## Why learn Maths?

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
sentences = [
    "The iconic red buses are the main mode of transport for CUETians.",
    "CUET students commute from Chittagong city on the university's red buses.",
    "Linear algebra and multivariable calculus are core maths for ML."
]
```

```python
embeddings = model.encode(sentences)
embeddings
```

```python
embeddings[0] @ embeddings[1]
```
